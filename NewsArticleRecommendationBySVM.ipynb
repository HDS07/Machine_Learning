```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics.pairwise import cosine_similarity
```

```python
news_data = pd.read_csv('/content/newsdataset.csv')
news_data.fillna('', inplace=True)
```

```python
news_data['content'] = news_data['title'] + " " + news_data['summary'] + " " + news_data['keywords']
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf_vectorizer.fit_transform(news_data['content'])
```

```python
svm_model = SVC(kernel='linear', probability=True)
svm_model.fit(tfidf_matrix, news_data['id'])
```

```python
def recommend_articles(user_keywords, model, tfidf_vectorizer, tfidf_matrix, news_data, top_n=5):
    # Transform the user input keywords to match TF-IDF matrix
    user_vector = tfidf_vectorizer.transform([user_keywords])

    # Calculate cosine similarity between user vector and all articles in the dataset
    similarity_scores = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # Get the indices of the top_n most similar articles
    similar_indices = similarity_scores.argsort()[-top_n:][::-1]

    # Retrieve and display recommended articles
    recommended_articles = news_data.iloc[similar_indices]
    pd.set_option('display.max_colwidth', None)

    for i, row in recommended_articles.iterrows():
        print(f"Title: {row['title']}")
        print(f"Summary: {row['summary']}")
        print(f"Link: {row['link']}\n")
        print("=" * 80)

    # Reset display options to default after printing
    pd.reset_option('display.max_colwidth')
```

```python
user_keywords = "Vodafone"
recommend_articles(user_keywords, svm_model, tfidf_vectorizer, tfidf_matrix, news_data, top_n=5)
```
